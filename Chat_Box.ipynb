## Preguntando a un DataFrame de Pandas

```python
import pandas as pd

# Import dataset
df = pd.read_csv('/home/chickenux/Documents/Chatbox/vgsales.csv')

# Keep only first 100 rows
df_subset = df.head(100)

# Convert those 100 rows to string
df_string = df_subset.to_string()

# Questions
QE = ['Which publisher has sold the most games globally?',
     'What are the top 5 platforms with the highest total sales worldwide?',
     'In which region are Japanese games (Publisher or Platform from Japan) most successful?']
QS = [
    '¿Qué distribuidor ha vendido más juegos a nivel mundial?',
    '¿Cuáles son las 5 plataformas con mayores ventas totales en todo el mundo?',
    '¿En qué región tienen más éxito los juegos japoneses (por editor o plataforma de Japón)?'
]


# Pass parameters
client = OpenAI()

for question in QS:
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {
                "role": "system",
                "content": (
                    "You are a game assistant. Your job is to answer questions, "
                    "recommend games, and inform the client about them. "
                    "Here are the first 100 records of the dataset:\n" + df_string
                )
            },
            {"role": "user", "content": question}
        ]
    )

    # Obtain answer with API
    answer = response.choices[0].message.content
    print(answer + "\n")
```
